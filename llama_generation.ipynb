# Testy z LLaMA 3.3 (Groq API)

```python
import pandas as pd
import numpy as np
CZY_GENEROWAC = False
df = pd.read_csv('prompts2.csv', sep=';')
print(df['Prompt'].dtype, df['Flag'].dtype)
```

```python
import os
from dotenv import load_dotenv

load_dotenv('tokens.env')
klucz_llama = os.getenv('TOKEN_GROQ')

if klucz_llama is None:
    print("Brak klucza API w pliku tokens.env")
else:
    print(f"Klucz API zaladowany")
```

```python
from groq import Groq

client = Groq(api_key=klucz_llama)

testowa_odpowiedz = client.chat.completions.create(
    model="llama-3.1-8b-instant",
    messages=[{"role": "user", "content": "czy działasz? odpowiedz jednym słowem"}],
    temperature=0.5,
    max_tokens=50
)
print(f"Model działa: {testowa_odpowiedz.choices[0].message.content}")
```

```python
from llama_processing import DataModel
```

```python
data_model = DataModel(df, client, rate_limit=28, model="llama-3.1-8b-instant")
data_model.make_prompts()
data_model()
```

```python
if CZY_GENEROWAC:
    responses_df = data_model.prompts()
    data_model.save_results()
else:
    data_model.load_results()
    responses_df = data_model.responses
```

```python
metrics_df, var_df = data_model.compute_accuracy_and_variance()
metrics_df.to_csv('saved_responses_llama/metrics.csv', index=False, encoding='utf-8-sig')
var_df.to_csv('saved_responses_llama/controversial.csv', index=False, encoding='utf-8-sig')
print("Zapisano metryki")
```

```python
results_table = metrics_df.copy()
results_table['Prefix'] = results_table['Strategy'].map(lambda x: data_model.strategies.get(x, ''))
results_table = results_table.set_index('Strategy').sort_values('Accuracy', ascending=False)
results_table
```
